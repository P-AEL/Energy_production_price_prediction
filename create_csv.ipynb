```python
import pickle 
import pandas as pd
```

```python
df = pd.read_pickle("merged_dataset_prep_utc.pkl")
```

```python
df.columns.values
```

```python
substrings_solar = ["CloudCover", "Radiation", "Temperature_Point"]
substrings_demand = ["WindSpeed_Point", "WindDirection_Point", "RelativeHumidity_Point", "demand", "Total", "pes10"]
substrings_energy_data = ["dtm", "MIP", "Solar_MW", "Solar_capacity_mwp", "Solar_installedcapacity_mwp", "Wind_MW", "SS_Price", "boa_MWh", "DA_Price", "Wind_MWh_credit", "Solar_MWh_credit"]
```

```python
filtered_columns_solar = [col for col in df.columns if any(sub in col for sub in substrings_solar)]
solar_df = df[["valid_time", "reference_time"] + filtered_columns_solar + substrings_energy_data]

filtered_columns_demand = [col for col in df.columns if any(sub in col for sub in substrings_demand)]
demand_df = df[["valid_time", "reference_time"] + filtered_columns_demand + substrings_energy_data]


solar_demand_cols = set(filtered_columns_solar + filtered_columns_demand + substrings_energy_data)
wind_cols = [col for col in df.columns if col not in solar_demand_cols and col not in ["valid_time", "reference_time"]]
wind_df = df[["valid_time", "reference_time"] + wind_cols + substrings_energy_data]
```

```python
solar_df.columns.values
```

```python
solar_df = solar_df.drop(columns= ['Temperature_Point0_dwd_demand', 'Temperature_Point1_dwd_demand',
       'Temperature_Point2_dwd_demand', 'Temperature_Point3_dwd_demand',
       'Temperature_Point4_dwd_demand', 'Temperature_Point5_dwd_demand',
       'Temperature_Point6_dwd_demand', 'Temperature_Point0_ncep_demand',
       'Temperature_Point1_ncep_demand', 'Temperature_Point2_ncep_demand',
       'Temperature_Point3_ncep_demand', 'Temperature_Point4_ncep_demand',
       'Temperature_Point5_ncep_demand', 'Temperature_Point6_ncep_demand'])
```

```python
solar_df.columns = solar_df.columns.str.replace("_pes10", "")
```

```python
demand_df = demand_df.drop(columns= ['Temperature_Point0_dwd_pes10',
       'Temperature_Point1_dwd_pes10', 'Temperature_Point2_dwd_pes10',
       'Temperature_Point3_dwd_pes10', 'Temperature_Point4_dwd_pes10',
       'Temperature_Point5_dwd_pes10', 'Temperature_Point6_dwd_pes10',
       'Temperature_Point0_ncep_pes10', 'Temperature_Point1_ncep_pes10',
       'Temperature_Point2_ncep_pes10', 'Temperature_Point3_ncep_pes10',
       'Temperature_Point4_ncep_pes10', 'Temperature_Point5_ncep_pes10',
       'Temperature_Point6_ncep_pes10',])
```

```python
demand_df.columns = demand_df.columns.str.replace("_demand", "")
```

```python
solar_df.to_csv("solar1.csv")
demand_df.to_csv("demand1.csv")
```
